```python
#content based recommendation
df=movies.merge(credits,on='title')
df.shape
```

```python
df=df[['genres','id','keywords','overview','popularity','title','cast','vote_average','vote_count','crew']]
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['genres'][0]
df['keywords'][0]
df['cast'][0]
df['crew'][1]
```

```python
import ast
def convert(element):
    l=[]
    for i in ast.literal_eval(element):
        l.append(i["name"])
    return l

def convertCast(element):
    l=[]
    count=1
    for i in ast.literal_eval(element):
        if count==4:
            break
        
        l.append(i["name"])
        count=count+1
    return l


def convertCrew(element):
    l=[]
    for i in ast.literal_eval(element):
        if i['job']=="Director of Photography":
          l.append(i['name'])
    return l

    

```

```python
df['genres']=df['genres'].apply(convert)

df['keywords']=df['keywords'].apply(convert)
df['cast']=df['cast'].apply(convertCast)
df['crew']=df['crew'].apply(convertCrew)
df.head()

```

```python
df['overview']=df['overview'].apply(lambda x:x.split())
df.head()

                     
```

```python
df['genres']=df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew']=df['crew'].apply(lambda x:[i.replace(" ","") for i in x])

```

```python
#now concatenate
df['recommendKeyWords']=df['genres']+df['keywords']+df['cast']+df['crew']+df['overview']
new_df=df[['id','title','recommendKeyWords']]
new_df.head()

```

```python
new_df['recommendKeyWords']=new_df['recommendKeyWords'].apply(lambda x:" ".join(x))
```

```python
#convert into lowercase(recommended)
new_df['recommendKeyWords']=new_df['recommendKeyWords'].apply(lambda x:x.lower())
new_df.head()
new_df['recommendKeyWords'][0]
```

```python
#remove repeated words like action,actions and others as well
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(element):
    l=[]
    for i in element.split():
        l.append(ps.stem(i))
    return " ".join(l)

new_df['recommendKeyWords']=new_df['recommendKeyWords'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english") 
vectors=cv.fit_transform(new_df['recommendKeyWords']).toarray()
vectors.dtype
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)

```

```python
def recommend(movie):
    movie_index=int(new_df[new_df['title']==movie].index.values[0])
    distance=list(enumerate(similarity[movie_index]))
    sorted_list=sorted(distance,reverse=True,key=lambda x:x[1])[1:6]# 0 indexed movie is the 
    #same whose related movies we want
    cleaned=[(i,float(score)) for i,score in sorted_list]
    for i in cleaned:
        print(new_df.iloc[i[0]]['title'])

    
    
```
